```python
import pandas as pd
import random
import yfinance as yf
import matplotlib.pyplot as plt

assignments = pd.read_csv('student_assignments.csv')

problem_tickers = set()

# random_students = random.sample(range(1, 121), 10)
random_students = range(1, 121)

for student_id in random_students:
    # print(f"Student ID: {student_id}")
    student_data = assignments[assignments['Student'] == student_id]
    tickers = student_data.iloc[0, 1:].tolist()
    
    for ticker in tickers:
        try:
            stock = yf.Ticker(ticker)
            hist = stock.history(period="10y")
            
            if hist.empty:
                print(f"Ticker {ticker} is invalid or has no data.")
                problem_tickers.add(ticker)
                continue
            # plt.figure(figsize=(10, 6))
            # plt.plot(hist.index, hist['Close'], label=f"{ticker} - Close Price")
            # plt.title(f"Price History for {ticker} (Last 10 Years)")
            # plt.xlabel("Date")
            # plt.ylabel("Close Price")
            # plt.legend()
            # plt.grid()
            # plt.show()
        
        except Exception as e:
            print(f"Error fetching data for ticker {ticker}: {e}")

print(problem_tickers)
```
